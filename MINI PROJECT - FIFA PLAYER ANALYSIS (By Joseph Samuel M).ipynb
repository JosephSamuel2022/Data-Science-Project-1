## Data Preprocessing

### Dropping/Replacing Unwanted Columns

```python
df=df.drop("ID",axis=1)#ID is not a meaningful feature , ID does not affect the output(I considered 'Overall' as output)
```

```python
df = df.iloc[: , 1:]#Drops first column -> Unnamed column not necessary
```

```python
df.columns[df.isna().any()]
```

```python
cols=['Club', 'International Reputation','Skill Moves', 'Work Rate', 'Body Type', 'Real Face', 
       'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until',
         'LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
       'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM',
       'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 
        'Release Clause']
```

```python
df=df.drop(cols,axis=1)#Columns that are of no use in analysing the data are removed
```

```python
df
```

```python
df.columns
```

```python
cols2=['Photo','Flag','Club Logo','Weak Foot']
```

```python
df=df.drop(cols2,axis=1)#Columns that are of no use in analysing the data are removed
```

```python
df
```

```python
df.Height
```

```python
#Coverting height from string to floating point number
Height2=[]
for i in df['Height']:
    try:
        x=float(str(i)[0])+0.0833333*float(str(i)[2:])
        Height2.append(x)
    except:
        Height2.append(np.NaN)
                                      
```

```python
df["Height"]=Height2
```

```python
df['Height']
```

```python
df
```

### Handling NULL Values

```python
for j in df.columns:
    for i in df[j].unique():
        if i =='?' or i=='N/A':
            print(i)
else:
    print("No ambiguous null values")
#we can see that there are no ambiguous null values
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)#dropping the rows with null values
```

## BI VARIATE ANALYSIS

### 1.AGE vs OVERALL

```python
sb.jointplot(x=df["Age"],y=df["Overall"])
plt.show()
#Generally players within the range of 20-30 play better than very young/very old people(with some exceptions -> the two best players are of age 31 and 33)
```

```python
#For more colourful representation
sb.FacetGrid(df,hue='Age',height=10).map(sb.scatterplot,'Age','Overall',edgecolor='w').add_legend()
plt.show()
```

### AGGRESSION VS OVERALL

### MULTIVARIATE ANALYSIS

```python
df.corr()#correlation can be -1,1(or close to) ->good 0(or close to)->no correlation
```

```python
sb.heatmap(df.corr())#We can visualize the correlation for all columns , with all columns
plt.show()
#All GK(Goal Keeper) skills are interlinked(All goalkeepers in this data set have all goal keeping skills)
```

```python
sb.pairplot(data=df,x_vars=['Height','Vision','Overall'],y_vars='Overall',diag_kind=None)
plt.show()
#Players near the height 6 foot(5.7-6.2) generally play better than those people outside this range
#Generally players with good vision play better than others
```

```python
sb.pairplot(data=df,x_vars=['LongPassing','BallControl','Stamina'],y_vars='Overall',diag_kind=None)
plt.show()
#Generally players who have good LongPassing skill and very good BallControlling ability,are better players than those who have not mastered those skills
#We cannot infer the same for Stamina, there is also a possibility where some players can have less stamina and still be a better players
#Reason: The player might have a position where it requires less stamina , and he does his role very well(generally Central Defender or Defensive Midfielder)
```
